In this code, a scalar PDE is simulated. This is a toy problem to help Shilpa understand PDE system modeling within Pyomo and DoE within Pyomo.DoE. The system is modeled using:
$$\frac{\partial T(x,t)}{\partial t} = \alpha \frac{\partial^2 T(x,t)}{\partial x^2};~0<x<L,~t>0$$

Boundary conditions:
$$-k\frac{\partial T(0,t)}{\partial x} = u(t)$$

$$-k\frac{\partial T(L,t)}{\partial x} = h(T(L,t) - T_{c})$$

Initial condition:$$ T(x,0) = T_0$$

Measurement: $$ y(t) = T(0.5,t) + v(t)$$
$$v$$ is the measurement noise 

Unknown Parameter to estimate: $$\theta = \alpha$$

Known constants: $$L =1,~T_0 = 1,~T_c = 0,~k=1,~h = 1$$


Toy problem developed by ChatGPT and coded by Shilpa with the help of ChatGPT




```python
from pyomo.dae import ContinuousSet, DerivativeVar, Simulator
import numpy as np
from pathlib import Path
from metrics_logger import log_benchmark
from pyomo.contrib.parmest.experiment import Experiment
import pyomo.environ as pyo

## Defining the Pyomo model

class scalarPDE(Experiment):
    def __init__(self, data, nfe_t, ncp_t, nfe_x, ncp_x):
        """
        Arguments
        ---------
        data: object containing vital experimental information
        nfe: number of finite elements
        ncp: number of collocation points for the finite elements
        """
        self.data = data
        self.nfe_t = nfe_t # Number of finite time elements
        self.nfe_x = nfe_x # Number of finite space elements
        self.ncp_x = ncp_x
        self.ncp_t = ncp_t
        self.model = None

        #############################
        # End constructor definition

    def get_labeled_model(self):
        if self.model is None:
            self.create_model()
            self.finalize_model()
            self.label_experiment()
        return self.model

    ## Creating Model

    def create_model(self):
        # Define the model
        m = self.model = pyo.ConcreteModel()

        # Discretize the time and state space

        m.t = ContinuousSet(bounds=[0, 1]) # Need to discretize time space
        m.x_space = ContinuousSet(bounds = [0,1]) # Need to discretize state space

        # Define variables

        m.T = pyo.Var( m.x_space, m.t, within=pyo.Reals) # Varies with time and space
        m.u = pyo.Var(m.t, within=pyo.Reals)

        # Unknown parameter
        m.alpha = pyo.Var(within=pyo.Reals)

        # Define constants

        m.h = pyo.Param(initialize=self.data["h"], mutable=False)
        m.T_c = pyo.Param(initialize=self.data["Tc"], mutable=False)
        m.T_0 = pyo.Param(initialize=self.data["T0"], mutable=False)

        # Differential variables wrt x and t
        m.dTdt = DerivativeVar(m.T, wrt=m.t)
        m.dTdx = DerivativeVar(m.T, wrt=m.x_space)
        m.d2Tdx2 = DerivativeVar(m.dTdx, wrt=m.x_space) ## Second order derivative wrt x

        ## Define PDE

        @m.Constraint(m.x_space,m.t)
        def _pde_equation(m,x_space,t):
            x_init = m.x_space.first()
            x_final = m.x_space.last()

            ## If at either ends, enforce boundary conditions, so skip PDE

            if x_space == x_init or x_space == x_final:
                return pyo.Constraint.Skip
            else:
                return (m.dTdt[x_space,t] == m.alpha*m.d2Tdx2[x_space,t])

        ## Define boundary conditions

        x_left = m.x_space.first()
        x_right = m.x_space.last()
        
        @m.Constraint(m.t)
        def left_bc(m, t):
            # Neumann BC at x = 0:  dT/dx(0,t) = -u(t)
            return m.dTdx[x_left, t] == -m.u[t]
        
        @m.Constraint(m.t)
        def right_bc(m, t):
            # Convective BC at x = L: dT/dx(L,t) = -h * ( T(L,t) - T_c )
            return m.dTdx[x_right, t] == -m.h * (m.T[x_right, t] - m.T_c)


        


    ## Finalizing model

    def finalize_model(self):
        m = self.model

        # Unpacking data before simulation
        control_points = self.data["control_points"]
        t_range = self.data["t_range"]                    # list/array of time points
        u_bounds = self.data["u_bounds"]                  # (umin, umax)
        alpha_nominal = self.data["alpha_nominal"]        # scalar nominal diffusivity
        # -----------------------------------------
        # 2. Set up spatial and temporal domains
        # -----------------------------------------
        # (a) Make sure the spatial ContinuousSet includes required sensor locations, etc.
        #     e.g. x = 0.5 must exist so that m.T[0.5, t] is addressable.
        if x_range is not None:
            for xval in x_range:
                m.x_space.update([xval])
    
        # (b) Update the time ContinuousSet with the full simulation horizon
        #     and all the control change times.
        for tval in t_range:
            m.t.update([tval])
        for t_ctrl in control_points:
            m.t.update([t_ctrl])
    
        # Save the control times (this mirrors `m.t_control = control_points`)
        # Use a Pyomo Set so we can iterate in constraints.
        m.t_control = pyo.Set(initialize=list(control_points.keys()), ordered=True)
    
        # -----------------------------------------
        # 3. Fix the unknown physical parameter(s)
        # -----------------------------------------
        # For DoE, alpha is treated as "true but unknown".
        # We fix it to a nominal value here so sensitivities can be computed.
        m.alpha.fix(alpha_nominal)
    
        # -----------------------------------------
        # 4. Apply bounds to design variables
        # -----------------------------------------
        # (a) Initial condition decision variable: m.T0
        #     In the ODE code you did:
        #         m.x[0].setlb(...); m.x[0].setub(...)
        #     Here, m.T0 already had physical bounds when it was declared.
        #     If you instead stored bounds in data, you can set them now:
        if "T0_bounds" in self.data:
            Tmin, Tmax = self.data["T0_bounds"]
            m.T0.setlb(Tmin)
            m.T0.setub(Tmax)
    
        # (b) Control input bounds for the boundary flux u(t)
        umin, umax = u_bounds
        for t in m.t:
            m.u[t].setlb(umin)
            m.u[t].setub(umax)
    
        # -----------------------------------------
        # 5. Initialize control input profile m.u[t]
        # -----------------------------------------
    
        for t in m.t:
            if t in control_points:
                # direct guess provided
                m.u[t].value = control_points[t]
            else:
                # use the most recent earlier control point's guess
                prev_ctrl_times = [tc for tc in control_points if tc <= t]
                if len(prev_ctrl_times) == 0:
                    first_tc = min(control_points.keys())
                    m.u[t].value = control_points[first_tc]
                else:
                    last_tc = max(prev_ctrl_times)
                    m.u[t].value = control_points[last_tc]
    
        # -----------------------------------------
        # 6. Enforce piecewise-constant control
        # -----------------------------------------
        # In your ODE code:
        #   @m.Constraint(m.t - control_points) ...
        # We do the same: for any time that's NOT itself a control point,
        # force u(t) == u(previous control time).
        @m.Constraint(m.t - m.t_control)
        def u_control(m, t):
            """
            Piecewise constant boundary heat flux between control points
            """
            prev_tc = max(tc for tc in m.t_control if tc < t)
            return m.u[t] == m.u[prev_tc]
    
        m.u_control = u_control
    
        # -----------------------------------------
        # 7. Discretize the PDE
        # -----------------------------------------
        # ODE version only discretized in time.
        # PDE version must discretize in space *and* time.
        #
        # Standard workflow:
        #   1. Discretize space first (to convert PDE -> system of ODEs)
        #   2. Then discretize time (to convert ODEs -> algebraic constraints)
        #
        # We'll assume the class stored nfe_x, ncp_x for space,
        # and nfe_t, ncp_t for time.
        space_disc = pyo.TransformationFactory("dae.collocation")
        space_disc.apply_to(
            m,
            wrt=m.x_space,
            nfe=self.nfe_x,
            ncp=self.ncp_x,
        )
    
        time_disc = pyo.TransformationFactory("dae.collocation")
        time_disc.apply_to(
            m,
            wrt=m.t,
            nfe=self.nfe_t,
            ncp=self.ncp_t,
        )




## Labeling the model

    def label_experiment(self):
        m = self.model

        # Tell Pyomo.DOE which variables are experimentally observed
        m.experiment_outputs = pyo.Suffix(direction=pyo.Suffix.LOCAL)

        # There's only one sensor, at x_s = 0.5
        x_s = 0.5

        # Register T(0.5, t) as a measured output at each measurement time t in m.t_control
        for t_m in m.t_control:
        m.experiment_outputs[m.T[x_s, t_m]] = None

        # Tell Pyomo.DOE what the measurement error is for each of those observations
        m.measurement_error = pyo.Suffix(direction=pyo.Suffix.LOCAL)
        meas_error = 1e-2  # same idea as in your ODE code

        for t_m in m.t_control:
        m.measurement_error[m.T[x_s, t_m]] = meas_error


        ## Define initial conditions

        t0 = m.t.first()
        T0_val = self.data["T0"]

        for x in m.x_space:
             m.T[x, t0].fix(T0_val)

        # 1. Tell Pyomo.DOE which things are experimental inputs (design variables)
        m.experiment_inputs = pyo.Suffix(direction=pyo.Suffix.LOCAL)

        # (a) Initial uniform temperature T0 is a design choice.
        #     This plays the same conceptual role as m.x[m.t.first()] in your ODE model.
        m.experiment_inputs[m.T0] = None

        # (b) Boundary heat flux u(t) at each control time is a design choice.
        #     This plays the same role as m.u1[t] in your ODE model.
        for t_ctrl in m.t_control:
            m.experiment_inputs[m.u[t_ctrl]] = None


        # 2. Tell Pyomo.DOE which parameters are unknown and to be estimated
        m.unknown_parameters = pyo.Suffix(direction=pyo.Suffix.LOCAL)

        # Here, the only unknown physical parameter we care about is alpha (thermal diffusivity).
        # We store its nominal (fixed) value as the suffix value.
        m.unknown_parameters[m.alpha] = pyo.value(m.alpha)

            
            
            

```
