This code builds on the transient heat conduction example from the Pyomo tutorial examples by Jeff Kantor. The model simulates the following equation:

$\frac{\partial T}{\partial t} = \alpha \frac{ \partial ^2 T}{\partial x^2}$

$\alpha = 1$

$T(0,x) = 0$ for all $0 \leq x \leq 1$

$T(t,1) = 1$ for all $t >0$

$\frac{dT}{dx}(t,0) = 0$ for all $t \geq 0$



```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D ## Needed to create 3D plots

import shutil ## Shell utilities for python, allows to copy/move/rename files or delete directories
import sys ## interact with python interpreter itself
import os.path ## path manipulation utilities


from pyomo.environ import *
from pyomo.dae import *
from idaes.core.util import DiagnosticsToolbox

m = ConcreteModel()

m.x = ContinuousSet(bounds = (0,1))
m.t = ContinuousSet(bounds =(0,2))

m.T = Var(m.t,m.x)

m.dTdt = DerivativeVar(m.T, wrt = m.t)

m.dTdx = DerivativeVar(m.T, wrt = m.x)

m.d2Tdx2 = DerivativeVar(m.dTdx, wrt = m.x) ## This is how pyomo knows to implement the boundary condition, if
# m.d2Tdx2 = DerivativeVar(m.T, (wrt = m.x,m.x)), then pyomo may not know how to implement the Neumann condition

@m.Constraint(m.t,m.x)

def pde(m,t,x):
    if x == 0 or t == 0:
        return Constraint.Skip
    else:
        return m.dTdt[t,x] == m.d2Tdx2[t,x]

m.obj = Objective(expr =1)

## Initial condition

m.ic = Constraint(m.x,rule = lambda m, x: m.T[0,x] ==0 if x<1 else Constraint.Skip)

## Boundary conditions

m.bc1 = Constraint(m.t, rule = lambda m, t: m.T[t,1] == 1)

m.bc2 = Constraint(m.t, rule = lambda m, t: m.dTdx[t,0] == 0)

## initial condition is at t = 0 and boundary conditions are at x = 0, so backwards methods may be more appropriae

# TransformationFactory("dae.finite_difference").apply_to(m,nfe = 20,scheme = "BACKWARD", wrt = m.x)
# TransformationFactory("dae.finite_difference").apply_to(m,nfe = 50,scheme = "BACKWARD", wrt = m.t)

TransformationFactory("dae.collocation").apply_to(m,nfe = 20,ncp = 10, wrt = m.x)
TransformationFactory("dae.collocation").apply_to(m,nfe = 50,ncp = 10, wrt = m.t)

import idaes 
from idaes.core.util.model_diagnostics import degrees_of_freedom
print("Degrees of freedom:", degrees_of_freedom(m))

dt = DiagnosticsToolbox(m)
dt.report_structural_issues()
dt.display_underconstrained_set()
dt.display_overconstrained_set()




```
