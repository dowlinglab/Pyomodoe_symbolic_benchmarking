In this code, a PDE model of a PFR is simulated. This replicates Waldron et al (2000/2019(?)) paper(s). PFR with an esterification reaction is considered. Benzoic acid (BA) + Ethanol in equilibrium with ethyl benzoate and water. The model for the reactor:

$$ r_{BA} = - k C_{BA}$$



The reaction constant:

$$ k = exp\bigg( -KP_1 + \frac{KP_2\times 10000}{R}\bigg) \bigg[ \frac{1}{T} - \frac{1}{T_m}\bigg]$$

$$ k_0 = -KP_1 + \frac{KP_2 \times 10000}{R T_m} $$

$$ E_A = KP_2 \times 10000$$


Balance:

$$\frac{\partial C_{BA}}{\partial t} = -\nu(t) \frac{\partial C_{BA}}{\partial v} + r_{BA}$$




```python
from pyomo.dae import ContinuousSet, DerivativeVar
import pyomo.environ as pyo
from pyomo.contrib.doe import DesignOfExperiments
import numpy as np


class waldronPDE:
    """
    1D diffusion PDE with:
    - left boundary heat flux control u(t)
    - right boundary insulated (no flux)
    - uniform initial temperature T0
    Structured so DesignOfExperiments can call get_labeled_model().
    """

    def __init__(self, data, nfe_t, ncp_t, nfe_x, ncp_x):
        self.data   = data
        self.nfe_t  = nfe_t
        self.ncp_t  = ncp_t
        self.nfe_z  = nfe_z
        self.ncp_z  = ncp_z
        self.model  = None

    def get_labeled_model(self):
        if self.model is None:
            self._create_model()
            self._finalize_model()
            self._label_experiment()
        return self.model

    def _create_model(self):
        m = self.model = pyo.ConcreteModel()

        # pull time horizon from data
        t_lo = min(self.data["t_range"])
        t_hi = max(self.data["t_range"])

        # domains
        m.t = ContinuousSet(bounds=[t_lo, t_hi])
        m.z_space = ContinuousSet(bounds=[0.0, 2.0]) ## Length of the reactor in m = 2.0 meter

        # fields
        m.CBA = pyo.Var(m.z_space, m.t, within=pyo.Reals)   # Concentration of Benzoic acid
        m.T = pyo.Var( m.t, within=pyo.Reals)   # Temperature of the reaction mixture

        ## Design variables

        m.v0 = pyo.Var(within=pyo.Reals) ## Initial ramp flow rate m^3/s
        m.CBA0 = pyo.Var(within=pyo.Reals) ## Initial benzoic acid flow rate m^3/s
        m.T0 = pyo.Var(within=pyo.Reals) ## Initial temperature K
        m.alphav = pyo.Var(within=pyo.Reals) 
        m.alphaT = pyo.Var(within=pyo.Reals)
        

        # Parameters to estimate
       
        m.KP1 = pyo.Var(within=pyo.Reals)
        m.KP2 = pyo.Var(within=pyo.Reals)

        # Defining fixed parameters
        m.dia   = pyo.Param(initialize=self.data["dia"],  mutable=False) # Reactor diameter in meter
        m.len = pyo.Param(initialize=self.data["len"], mutable=False) # Reactor length in meter
        m.R = pyo.Param(initialize=self.data["R"], mutable=False) # Real gas constant J/molK
        m.Tm = pyo.Param(initialize=self.data["Tm"], mutable=False)  # Mean temperature K

        # time and space derivatives
        m.dCBAdt   = DerivativeVar(m.CBA, wrt=m.t)
        m.dCBAdz   = DerivativeVar(m.CBA, wrt=m.z_space)
        
        # interior PDE (skip boundaries in x)
        @m.Constraint(m.z_space, m.t)
        def pde_eqn(m, z, t):
            xL = m.z_space.first()
            xR = m.z_space.last()
            if (z == xL) or (z == xR):
                return pyo.Constraint.Skip
            return m.dCBAdt[z, t] == (m.v0 - m.alphav*t)/ (np.pi*m.len* (m.dia/2)**2)*m.dCBAdz[z,t]\
            - np.exp(-m.KP1 - ( (m.KP2*10000)/(m.R)* (1/m.T[t] - 1/m.Tm) ))


        ## Adding expression for flowrate ramp

        @m.constraint(m.t):
        def flow_ramp(m,t):
            return m.v[t] == m.v0 - m.alphav * t

        ## Adding expression for temperature ramp

        @m.constraint(m.t):
        def temp_ramp(m,t):
            return m.T[t] == m.T0 - m.alphaT * t


     

        # Adding boundary conditions in _finalize_model()

    def _finalize_model(self):
        m = self.model

        # -------------------
        # unpack data
        # -------------------
        control_points = self.data["control_points"]     
        t_range        = self.data["t_range"]           
        u_bounds       = self.data["u_bounds"]           
        alphav  = self.data["alphav"]      
        T0       = self.data.get("T0")  
        CBA_range = self.data["CBA_range"]
        alphaT = self.data["alphaT"]
        KP1 = self.data["KP1"]
        KP2 = self.data["KP2"]
        CBA0 = self.data["CBA0"]
        
        
        # -------------------
        # augment spatial grid before discretization
        # -------------------
        if x_range is not None:
            for xv in x_range:
                m.x_space.update([xv])

        # boundary nodes
        x_left_val  = min(list(m.x_space))
        x_right_val = max(list(m.x_space))

        # -------------------
        # augment time grid before discretization
        # -------------------
        for tt in t_range:
            m.t.update([tt])
        for tt in control_points.keys():
            m.t.update([tt])

        # control times as an ordered Set
        m.t_control = pyo.Set(
            initialize=list(control_points.keys()),
            ordered=True,
        )

        # -------------------
        # fix alpha to nominal
        # -------------------
        m.alpha.fix(alpha_nominal)

        # -------------------
        # bounds on T0
        # -------------------
        if "T0_bounds" in self.data:
            Tmin, Tmax = self.data["T0_bounds"]
            m.T0.setlb(Tmin)
            m.T0.setub(Tmax)

        # -------------------
        # bounds + init for u(t)
        # -------------------
        umin, umax = u_bounds
        for t in m.t:
            m.u[t].setlb(umin)
            m.u[t].setub(umax)

            # step-wise constant guess using last control point <= t
            prev_ctrl_times = [tc for tc in control_points if tc <= t]
            if len(prev_ctrl_times) == 0:
                first_tc = min(control_points.keys())
                m.u[t].value = control_points[first_tc]
            else:
                last_tc = max(prev_ctrl_times)
                m.u[t].value = control_points[last_tc]

        # enforce piecewise-constant control between control points
        @m.Constraint(m.t - m.t_control)
        def u_hold(m, t):
            prev_tc = max(tc for tc in m.t_control if tc < t)
            return m.u[t] == m.u[prev_tc]
        # do NOT overwrite m.u_hold = u_hold

        # -------------------
        # boundary conditions
        # -------------------

        # left: flux BC, matches the control
        @m.Constraint(m.t)
        def left_bc(m, t):
            return m.dTdx[x_left_val, t] == -m.u[t]

        # right: insulated (Neumann = 0), to be compatible with uniform T0
        @m.Constraint(m.t)
        def right_bc(m, t):
            return m.dTdx[x_right_val, t] == 0.0

        # -------------------
        # store special nodes
        # -------------------
        m.x_sensor     = pyo.Param(initialize=x_sensor_val,  mutable=False)
        m.x_left_node  = pyo.Param(initialize=x_left_val,    mutable=False)
        m.x_right_node = pyo.Param(initialize=x_right_val,   mutable=False)

        # -------------------
        # discretize space then time
        # -------------------
        disc_space = pyo.TransformationFactory("dae.collocation")
        disc_space.apply_to(
            m,
            wrt=m.x_space,
            nfe=self.nfe_x,
            ncp=self.ncp_x,
        )

        disc_time = pyo.TransformationFactory("dae.collocation")
        disc_time.apply_to(
            m,
            wrt=m.t,
            nfe=self.nfe_t,
            ncp=self.ncp_t,
        )

        # -------------------
        # initial condition T(x, t0) = T0
        # -------------------
        t0 = min(list(m.t))
        T0_val = pyo.value(m.T0)
        for x in m.x_space:
            m.T[t0].fix(T0_val)

    def _label_experiment(self):
        m = self.model

        # measured outputs
        m.experiment_outputs = pyo.Suffix(direction=pyo.Suffix.LOCAL)
        x_s = pyo.value(m.x_sensor)
        for t_m in m.t_control:
            m.experiment_outputs[m.T[x_s, t_m]] = None

        # measurement error
        m.measurement_error = pyo.Suffix(direction=pyo.Suffix.LOCAL)
        meas_err = 1e-2
        for t_m in m.t_control:
            m.measurement_error[m.T[x_s, t_m]] = meas_err

        # experiment inputs (design variables)
        m.experiment_inputs = pyo.Suffix(direction=pyo.Suffix.LOCAL)
        m.experiment_inputs[m.T0] = None
        for t_ctrl in m.t_control:
            m.experiment_inputs[m.u[t_ctrl]] = None

        # unknown parameter we're designing for
        m.unknown_parameters = pyo.Suffix(direction=pyo.Suffix.LOCAL)
        m.unknown_parameters[m.alpha] = pyo.value(m.alpha)


# --------------------------
# data for the PDE experiment
# --------------------------

# make ALL control points 0.0 to avoid forcing gradients at t≈0
flat_u = {
    0.0:   0.5,
    0.125: 0.5,
    0.25:  0.5,
    0.375: 0.5,
    0.5:   0,
    0.625: -0.5,
    0.75:  -0.2,
    0.875: -0.5,
    1.0:   -0.5,
}

## Defining all data to be used in the experiment


data_FR1 = {
    "tM,i Time the sample was measured (s)": [408, 828, 1248, 1667, 2084, 2504, 2923, 3342, 3763, 4182, 4603, 5024, 5444, 5864],
    "tIn,i Time the sample entered the reactor (s)": [110, 511, 910, 1304, 1693, 2080, 2460, 2833, 3198, 3549, 3885, 4199, 4481, 4722],
    "tL,i Time the sample left the reactor (s)": [314, 728, 1141, 1552, 1959, 2368, 2774, 3177, 3578, 3972, 4360, 4736, 5094, 5422],
    "τi Sample reaction time (s)": [204, 217, 231, 248, 267, 288, 314, 344, 380, 422, 474, 537, 613, 700],
    "X Conversion (%)": [5.8, 5.8, 6.4, 7.1, 7.1, 7.7, 8.3, 9.0, 9.6, 10.9, 12.2, 13.5, 14.7, 16.7],
    "Outlet Concentration Benzoic Acid (M)": [1.47, 1.47, 1.46, 1.45, 1.45, 1.44, 1.43, 1.42, 1.41, 1.39, 1.37, 1.35, 1.33, 1.30],
    "Outlet Concentration Ethyl Benzoate (M)": [0.09, 0.09, 0.10, 0.10, 0.11, 0.12, 0.13, 0.14, 0.15, 0.17, 0.18, 0.20, 0.23, 0.25]
}

data_FR2 = {
    "Time the sample was measured (t_M,i)": [152, 571, 991, 1411, 1830, 2251, 2670, 3090, 3510, 3931, 4350, 4770, 5190, 5610],
    "Time the sample entered the reactor (t_In,i)": [None, None, 17, 404, 787, 1170, 1549, 1925, 2298, 2668, 3033, 3393, 3749, 4098],
    "Time the sample left the reactor (t_L,i)": [None, 271, 680, 1089, 1496, 1904, 2309, 2714, 3117, 3520, 3920, 4318, 4715, 5109],
    "Sample reaction time (τ_i)": [623, 643, 663, 685, 709, 734, 760, 789, 819, 852, 887, 925, 966, 1010],
    "Conversion (X)": [44.9, 45.5, 46.2, 47.4, 48.1, 48.7, 50.0, 51.3, 51.9, 53.8, 55.1, 56.4, 57.7, 59.0],
    "Outlet Concentration Benzoic Acid (M)": [0.86, 0.85, 0.84, 0.82, 0.81, 0.80, 0.78, 0.76, 0.75, 0.72, 0.70, 0.68, 0.66, 0.64],
    "Outlet Concentration Ethyl Benzoate (M)": [0.70, 0.70, 0.72, 0.73, 0.74, 0.76, 0.78, 0.79, 0.81, 0.83, 0.86, 0.87, 0.90, 0.92]
}

data_FT = {
    "Time the sample was measured (s)": [419, 840, 1260, 1679, 2099, 2519, 2939, 3359, 3779, 4199, 4619, 5038, 5457, 5876],
    "Time the sample entered the reactor (s)": [np.nan, np.nan, 354, 739, 1124, 1505, 1883, 2258, 2629, 2996, 3358, 3714, 4064, 4407],
    "Time the sample left the reactor (s)": [np.nan, 150, 561, 971, 1379, 1786, 2193, 2599, 3003, 3406, 3807, 4206, 4602, 4995],
    "Sample reaction time (s)": [577, 596, 617, 639, 663, 688, 716, 745, 777, 811, 848, 888, 931, 979],
    "Reactor temperature at time tini (°C)": [139.2, 139.2, 136.0, 132.6, 129.1, 125.7, 122.3, 119.0, 115.7, 112.4, 109.1, 106.0, 102.8, 99.8],
    "Reactor temperature at time tLi (°C)": [139.2, 134.2, 130.5, 126.9, 123.2, 119.6, 115.9, 112.3, 108.7, 105.1, 101.6, 98.0, 94.5, 91.0],
    "X Conversion (%)": [41.5, 39.6, 35.1, 31.2, 26.7, 23.5, 19.7, 17.1, 14.5, 12.6, 10.0, 8.7, 7.5, 6.2],
    "Outlet Conc. Benzoic Acid (M)": [0.91, 0.94, 1.01, 1.07, 1.14, 1.19, 1.25, 1.29, 1.33, 1.36, 1.40, 1.42, 1.44, 1.46],
    "Outlet Conc. Ethyl Benzoate (M)": [0.64, 0.62, 0.55, 0.49, 0.42, 0.37, 0.31, 0.27, 0.22, 0.19, 0.16, 0.13, 0.11, 0.09]
}## Not sure what the nan values are: to ask Shuvo

# discretization
nfe_x = 5
ncp_x = 5
nfe_t = 8
ncp_t = 8


# wrapper class to satisfy DesignOfExperiments API
class ExperimentWrapper:
    def __init__(self, exp_obj):
        self._exp = exp_obj
    def get_labeled_model(self):
        return self._exp.get_labeled_model()


# build wrapped experiment
pde_expt = waldronPDE(
    data=data_pde,
    nfe_t=nfe_t,
    ncp_t=ncp_t,
    nfe_x=nfe_x,
    ncp_x=ncp_x,
)
wrapped_expt = ExperimentWrapper(pde_expt)

# DoE setup
fd_formula = "central"
step_size = 1e-3
objective_option = "determinant"
scale_nominal_param_value = True

doe_obj = DesignOfExperiments(
    wrapped_expt,
    fd_formula=fd_formula,
    step=step_size,
    objective_option=objective_option,
    scale_constant_value=1,
    scale_nominal_param_value=scale_nominal_param_value,
    prior_FIM=None,
    jac_initial=None,
    fim_initial=None,
    L_diagonal_lower_bound=1e-7,
    solver=pyo.SolverFactory("ipopt"),
    tee=False,
    get_labeled_model_args=None,
    _Cholesky_option=True,
    _only_compute_fim_lower=True,
)

# run DoE
doe_obj.run_doe()

# report results
print("Optimal experiment values:")
opt_design = doe_obj.results["Experiment Design"]

if len(opt_design) > 0:
    print(f"\tInitial temperature T0*: {opt_design[0]:.4f}")

if len(opt_design) > 1:
    u_list = opt_design[1:]
    print(
        "\tControl inputs u(t_ctrl)*: ["
        + ", ".join(f"{val:.4f}" for val in u_list)
        + "]"
    )

print("FIM at optimal design:")
print(np.array(doe_obj.results["FIM"]))

print("Objective value at optimal design:",
      pyo.value(doe_obj.model.objective))

print("Experiment Design Names:", doe_obj.results["Experiment Design Names"])
print("Solve time (s):", doe_obj.results["Solve Time"])
print("Build time (s):", doe_obj.results["Build Time"])
print("Initialization time (s):", doe_obj.results["Initialization Time"])
print("Total wall time (s):", doe_obj.results["Wall-clock Time"])
```
